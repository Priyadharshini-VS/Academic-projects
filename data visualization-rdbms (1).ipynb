```python
import pandas as pd
import seaborn as sns
import numpy as np
data=pd.read_csv("C:\\Users\\Sastr\\OneDrive\\Documents\\MARK.csv")
data.head(10)
#data.isna().sum())
```

```python
import matplotlib.pyplot as plt
plt.plot(data['MATHS'],data['ENGLISH'])
plt.xlabel("MATHS")
plt.ylabel("ENGLISH")
plt.title("LINE PLOT")
plt.show()
```
